```python
# ==========================================================
# CREDIT CARD FRAUD DETECTION
# TRAIN RANDOM FOREST (TOP 5 FEATURES)
# ==========================================================

import pandas as pd
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

from imblearn.over_sampling import SMOTE

# ==========================================================
# 1. LOAD DATASET
# ==========================================================

print("="*60)
print("CREDIT CARD FRAUD DETECTION")
print("="*60)

df = pd.read_csv("Dataset/creditcard.csv")

print("\nDataset berhasil dibaca.")

# ==========================================================
# 2. MEMILIH 5 FITUR
# ==========================================================

selected_features = [
    "Faktor Anomali Transaksi A",
    "Indikator Utama Penyimpangan",
    "Faktor Anomali Transaksi B",
    "Sensor Pola Kecurangan",
    "Skor Integritas Transaksi"
]

X = df[selected_features]

y = df["Class"]

print("\nFeature yang digunakan :")

print(selected_features)

# ==========================================================
# 3. TRAIN TEST SPLIT
# ==========================================================

X_train, X_test, y_train, y_test = train_test_split(

    X,
    y,

    test_size=0.20,

    random_state=42,

    stratify=y

)

print("\nData Training :", X_train.shape)

print("Data Testing :", X_test.shape)

# ==========================================================
# 4. SCALING
# ==========================================================

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)

print("\nScaling berhasil.")

# ==========================================================
# 5. SMOTE
# ==========================================================

print("\nSebelum SMOTE")

print(y_train.value_counts())

smote = SMOTE(random_state=42)

X_train, y_train = smote.fit_resample(

    X_train,

    y_train

)

print("\nSesudah SMOTE")

print(pd.Series(y_train).value_counts())

# ==========================================================
# 6. MEMBUAT MODEL
# ==========================================================

print("\nTraining Random Forest...")

model = RandomForestClassifier(

    n_estimators=100,

    random_state=42,

    n_jobs=-1

)

model.fit(

    X_train,

    y_train

)

print("Training selesai.")

# ==========================================================
# 7. PREDIKSI
# ==========================================================

y_pred = model.predict(X_test)

# ==========================================================
# 8. EVALUASI
# ==========================================================

accuracy = accuracy_score(y_test,y_pred)

precision = precision_score(y_test,y_pred)

recall = recall_score(y_test,y_pred)

f1 = f1_score(y_test,y_pred)

print("\n")

print("="*60)

print("HASIL EVALUASI")

print("="*60)

print(f"Accuracy  : {accuracy:.4f}")

print(f"Precision : {precision:.4f}")

print(f"Recall    : {recall:.4f}")

print(f"F1 Score  : {f1:.4f}")

print("\nClassification Report\n")

print(classification_report(

    y_test,

    y_pred

))

print("\nConfusion Matrix\n")

print(confusion_matrix(

    y_test,

    y_pred

))

# ==========================================================
# 9. SIMPAN MODEL
# ==========================================================

joblib.dump(

    model,

    "model/random_forest_model.pkl"

)

joblib.dump(

    scaler,

    "model/scaler.pkl"

)

print("\nModel berhasil disimpan.")

print("model/random_forest_model.pkl")

print("model/scaler.pkl")

print("="*60)
```
