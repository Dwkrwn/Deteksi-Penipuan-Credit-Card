```python
# ==========================================================
# PROJECT MACHINE LEARNING
# CREDIT CARD FRAUD DETECTION
# TAHAP 1 : EXPLORATORY DATA ANALYSIS (EDA)
# ==========================================================

# ===========================
# Import Library
# ===========================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Menampilkan seluruh kolom
pd.set_option('display.max_columns', None)

print("="*60)
print(" CREDIT CARD FRAUD DETECTION ")
print("="*60)

# ===========================
# Membaca Dataset
# ===========================

try:

    df = pd.read_csv("dataset/creditcard.csv")

    print("\nDataset berhasil dibaca.\n")

except FileNotFoundError:

    print("Dataset tidak ditemukan!")

    exit()

# ===========================
# Menampilkan 5 Data Pertama
# ===========================

print("\n===== 5 DATA PERTAMA =====\n")

print(df.head())

# ===========================
# Ukuran Dataset
# ===========================

print("\n===== UKURAN DATASET =====")

print(df.shape)

print("\nJumlah Baris :", df.shape[0])

print("Jumlah Kolom :", df.shape[1])

# ===========================
# Informasi Dataset
# ===========================

print("\n===== INFORMASI DATA =====")

print(df.info())

# ===========================
# Statistik Dataset
# ===========================

print("\n===== STATISTIK =====")

print(df.describe())

# ===========================
# Missing Value
# ===========================

print("\n===== MISSING VALUE =====")

print(df.isnull().sum())

# ===========================
# Duplikasi Data
# ===========================

print("\n===== DUPLIKAT =====")

print(df.duplicated().sum())

# ===========================
# Distribusi Class
# ===========================

print("\n===== DISTRIBUSI CLASS =====")

print(df["Class"].value_counts())

print("\nPersentase")

print(df["Class"].value_counts(normalize=True)*100)

# ===========================
# Visualisasi Distribusi Class
# ===========================

plt.figure(figsize=(6,5))

sns.countplot(x="Class", data=df)

plt.title("Distribusi Class")

plt.savefig("images/distribusi_class.png")

plt.show()

# ===========================
# Histogram Amount
# ===========================

plt.figure(figsize=(10,5))

plt.hist(df["Amount"], bins=50)

plt.title("Distribusi Amount")

plt.xlabel("Amount")

plt.ylabel("Jumlah")

plt.savefig("images/amount_histogram.png")

plt.show()

# ===========================
# Histogram Time
# ===========================

plt.figure(figsize=(10,5))

plt.hist(df["Time"], bins=50)

plt.title("Distribusi Time")

plt.xlabel("Time")

plt.ylabel("Jumlah")

plt.savefig("images/time_histogram.png")

plt.show()

# ===========================
# Korelasi
# ===========================

plt.figure(figsize=(18,15))

sns.heatmap(df.corr(), cmap="coolwarm")

plt.title("Correlation Matrix")

plt.savefig("images/correlation_matrix.png")

plt.show()

print("\nEDA SELESAI")

print("="*60)
```
