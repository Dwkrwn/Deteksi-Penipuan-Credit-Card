```python
# ==========================================================
# PROJECT MACHINE LEARNING
# CREDIT CARD FRAUD DETECTION
# TAHAP 2 : PREPROCESSING DATA
# ==========================================================

# ===========================
# Import Library
# ===========================

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from imblearn.over_sampling import SMOTE
import joblib

print("=" * 60)
print("PREPROCESSING DATA")
print("=" * 60)

# ===========================
# Membaca Dataset
# ===========================

df = pd.read_csv("dataset/creditcard.csv")

print("\nDataset berhasil dibaca.")

# ===========================
# Memisahkan Feature dan Target
# ===========================

X = df.drop("Class", axis=1)

y = df["Class"]

print("\nJumlah Feature :", X.shape[1])

print("Jumlah Data :", X.shape[0])

# ===========================
# Train Test Split
# ===========================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("\nTrain Test Split Berhasil")

print("Jumlah Data Training :", X_train.shape[0])

print("Jumlah Data Testing :", X_test.shape[0])

# ===========================
# Feature Scaling
# ===========================

scaler = RobustScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)

print("\nFeature Scaling Berhasil")

# ===========================
# SMOTE
# ===========================

print("\nSebelum SMOTE")

print(y_train.value_counts())

smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_train
)

print("\nSesudah SMOTE")

print(pd.Series(y_train_smote).value_counts())

# ===========================
# Menyimpan Data
# ===========================

joblib.dump(scaler, "model/scaler.pkl")

joblib.dump(
    (X_train_smote,
     X_test,
     y_train_smote,
     y_test),
    "model/preprocessing.pkl"
)

print("\nPreprocessing selesai.")

print("\nFile berhasil disimpan")

print("model/scaler.pkl")

print("model/preprocessing.pkl")

print("=" * 60)
```
